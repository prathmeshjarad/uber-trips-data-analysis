```python
#Importing Module
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
database=pd.read_csv(r"D:\Cyber_Success\Projects\Uber Trips Dataset Analysis\Excel Data\Clean data.csv")
```

```python
database.head(2)
```

```python
#Hotspot Location
plt.figure(figsize=(8,6))
sns.scatterplot(x="Lat",y="Lon",data=database)
plt.title("Hotspot Location")
plt.savefig(r"D:\Cyber_Success\Projects\Uber Trips Dataset Analysis\EDA Images\Hotspot Location.jpg")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Name of Day", data=database)
plt.title("Most and Least Pickup Day")
plt.xlabel("Day")
plt.ylabel("Number of Pickups")
plt.savefig(r"D:\Cyber_Success\Projects\Uber Trips Dataset Analysis\EDA Images\Most and Least Pickup Day.jpg")
plt.show()
```

```python
x = database["Day"].value_counts().index
y= database["Day"].value_counts().values
temp = pd.DataFrame({
    "Day":x,
    "Count of Pickup":y
})

plt.figure(figsize=(12,6))
sns.lineplot(x="Day", y="Count of Pickup", data=temp, marker="o")

plt.xticks([i for i in range(1,31,1)])
plt.title("Day-wise Pickup Trend")
plt.xlabel("Day of Month")
plt.ylabel("Number of Pickups")
plt.grid(axis="y")
plt.savefig(r"D:\Cyber_Success\Projects\Uber Trips Dataset Analysis\EDA Images\Day-wise Pickup Trend.jpg")
plt.show()
```

```python
x = database["AM/PM"].value_counts().index
y = database["AM/PM"].value_counts().values
plt.pie(
    y,
    labels=x,
    autopct="%.2f%%",
    radius=0.9
)
plt.title("AM / PM Pickup Trend")
plt.legend()
plt.savefig(r"D:\Cyber_Success\Projects\Uber Trips Dataset Analysis\EDA Images\AM_PM Pickup Trend.jpg")
plt.show()
```

```python
plt.figure(figsize=(13,6))

sns.histplot(
    x="Hour",
    data=database,
    bins=[i for i in range(0,24)]
)

plt.title("Most Pickup Time Slot (Hour-wise)")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Pickups")
plt.xticks([i for i in range(0,24)])
plt.grid(axis="y")
plt.savefig(r"D:\Cyber_Success\Projects\Uber Trips Dataset Analysis\EDA Images\Most Pickup Time Slot (Hour-wise).jpg")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(
    x="Name of Day",
    data=database,
    hue="AM/PM"
)
plt.title("Most Pickup vs Less Pickup Day (AM/PM wise)")
plt.ylabel("Count of Pickups")
plt.savefig(r"D:\Cyber_Success\Projects\Uber Trips Dataset Analysis\EDA Images\Most Pickup vs Less Pickup Day.jpg")
plt.show()
```

```python
# -------------------- STYLE SETTINGS --------------------
sns.set_style("whitegrid")
sns.set_context("talk")
sns.set_palette(["#1E88E5", "#FB8C00"])

# -------------------- KPI CALCULATIONS --------------------
total_trips = len(database)
busiest_day = database["Name of Day"].value_counts().idxmax()
peak_hour = database["Hour"].value_counts().idxmax()
ampm_ratio = database["AM/PM"].value_counts(normalize=True) * 100



# ================= Dashboard =================

# Create dashboard
fig, axes = plt.subplots(2, 3, figsize=(30, 16))
fig.suptitle("Uber Trips Dataset Analysis", fontsize=26, fontweight="bold")


# -------------------- BIG KPI CARDS --------------------
fig.text(
    0.06, 0.92,
    f"TOTAL TRIPS\n{total_trips}",
    fontsize=20,
    fontweight="bold",
    bbox=dict(
        facecolor="#E3F2FD",
        edgecolor="black",
        boxstyle="round,pad=1.2"
    )
)

fig.text(
    0.30, 0.92,
    f"BUSIEST DAY\n{busiest_day}",
    fontsize=20,
    fontweight="bold",
    bbox=dict(
        facecolor="#E8F5E9",
        edgecolor="black",
        boxstyle="round,pad=1.2"
    )
)

fig.text(
    0.65, 0.92,
    f"PEAK HOUR\n{peak_hour}:00",
    fontsize=20,
    fontweight="bold",
    bbox=dict(
        facecolor="#FFF3E0",
        edgecolor="black",
        boxstyle="round,pad=1.2"
    )
)

fig.text(
    0.90, 0.92,
    f"AM TRIPS\n{ampm_ratio['AM']:.1f}%\nPM TRIPS\n{ampm_ratio['PM']:.1f}%",
    fontsize=20,
    fontweight="bold",
    bbox=dict(
        facecolor="#FCE4EC",
        edgecolor="black",
        boxstyle="round,pad=1.2"
    )
)



# ================= Graph 1 =================
sns.scatterplot(
    x="Lat",
    y="Lon",
    data=database,
    ax=axes[0, 0],
    alpha=0.4,
    s=20
)
axes[0, 0].set_title("Hotspot Locations", fontsize=14)
axes[0, 0].set_xlabel("Latitude")
axes[0, 0].set_ylabel("Longitude")
axes[0, 0].grid(False)


# ================= Graph 2 =================
sns.countplot(
    x="Name of Day",
    data=database,
    ax=axes[0, 1],
    # order=database["Name of Day"].value_counts().index
)
axes[0, 1].set_title("Most and Least Pickup Day", fontsize=14)
axes[0, 1].set_xlabel("Day")
axes[0, 1].set_ylabel("Number of Pickups")
axes[0, 1].tick_params(axis="x", rotation=30)


# ================= Graph 3 =================
day_trend = database["Day"].value_counts().sort_index()

sns.lineplot(
    x=day_trend.index,
    y=day_trend.values,
    marker="o",
    color="#43A047",
    ax=axes[0, 2]
)
axes[0, 2].set_title("Day-wise Pickup Trend", fontsize=14)
axes[0, 2].set_xlabel("Day of Month")
axes[0, 2].set_ylabel("Number of Pickups")
axes[0, 2].grid(True)

# ================= Graph 4 =================
ampm_counts = database["AM/PM"].value_counts()

axes[1, 0].pie(
    ampm_counts.values,
    labels=ampm_counts.index,
    autopct="%1.1f%%",
    startangle=90,
    wedgeprops={"edgecolor": "black"}
)
axes[1, 0].set_title("AM / PM Pickup Distribution", fontsize=14)

# ================= Graph 5 =================
sns.histplot(
    data=database,
    x="Hour",
    bins=24,
    ax=axes[1, 1]
)
axes[1, 1].set_title("Most Pickup Time Slot (Hour-wise)", fontsize=14)
axes[1, 1].set_xlabel("Hour of Day")
axes[1, 1].set_ylabel("Number of Pickups")
axes[1, 1].set_xticks(range(0, 24))
axes[1, 1].grid(True)

# ================= Graph 6 =================
sns.countplot(
    x="Name of Day",
    hue="AM/PM",
    data=database,
    ax=axes[1, 2]
)
axes[1, 2].set_title("Pickup Comparison by Day (AM/PM)", fontsize=14)
axes[1, 2].set_xlabel("Day")
axes[1, 2].set_ylabel("Count of Pickups")
axes[1, 2].legend(title="Time")



# -------------------- LAYOUT FIX --------------------
plt.tight_layout(rect=[0, 0, 1, 0.88])


# -------------------- SAVE & SHOW --------------------
plt.savefig(
    r"D:\Cyber_Success\Projects\Uber Trips Dataset Analysis\EDA Images\Uber_Trips_Creative_Dashboard.png",
    dpi=300,
    bbox_inches="tight"
)


plt.show()
```
