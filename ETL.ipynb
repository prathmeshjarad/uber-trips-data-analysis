```python
#Changing Data Type
database['Date/Time']=database['Date/Time'].astype("datetime64[ns]")
```

```python
#InFo Extract Date Time
database['Name of Day']=database['Date/Time'].dt.day_name()
database['Day']=database['Date/Time'].dt.day
database['Month']=database['Date/Time'].dt.month
database['Year']=database['Date/Time'].dt.year
database['Hour']=database['Date/Time'].dt.hour
```

```python
def am_pm_create(hour):
    if 0 <= hour <= 12:
        return "AM"
    elif 12 < hour <= 23:
        return "PM"
database["AM/PM"] = database["Hour"].apply(am_pm_create)
database.head(4)
```

```python
#Export Data
database.to_csv(r"D:\Cyber_Success\Projects\Uber Trips Dataset Analysis\Excel Data\Clean data.csv",index=False)
```
